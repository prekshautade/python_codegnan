```python
#list comprehension
# syntax : expression for iterator in iterable if condition
n = 10
lst = []
for i in range (1,n+1):
    lst.append(i)
print(lst)    
```

```python
#list of n numbers using list comprehension
res=[x for x in range(1,n+1)]
print(res)
```

```python
n = int(input())
res=[x for x in range(1,n+1) if x%2==0]
print(res)
```

```python
#n even numbers
n = int(input())
res=[x*2 for x in range(n+1)]
print(res)
```

```python
#n odd numbers
n = int(input())
res=[x*2+1 for x in range(n)]
print(res)
```

```python
# n even numbers between the range
n1,n2 = map(int,input().split())
res = [x*2 for x in range(n1,n2)]
print(res)
```

```python
n1,n2 = map(int,input().split())
res = [x for x in range(n1,n2) if x%2==0]
print(res)
```

```python
# - List comprehension with if else
# - syntax : [true_exp if condition else false_exp for iterator in iterable]
```

```python
# list of numbers convert into 1's and 0's.
#even number - 0
#odd number - 1
lst = list(map(int,input().split()))
res = [0 if num%2==0 else 1 for num in lst]
print(res)
```

```python
# list of numbers convert into 1's and 0's.
#even number - leftshift by 1
#odd number - rightshift by 1
lst = list(map(int,input().split()))
res = [num << 1 if num%2==0 else num >>1 for num in lst]
print(res)
```

```python
# two dice combination
res = [(x,y) for x in range(1,7) for y in range(1,7)]
print(res)
```

```python
# sum of two dice faces is even
res = [(x,y) for x in range(1,7) for y in range(1,7) if (x+y)%2==0]
print(res)
```

```python
# two dice faces is same
res = [(x,y) for x in range(1,7) for y in range(1,7) if x == y]
print(res)
```

```python
# two dice face of 4
res = [(x,y) for x in range(1,7) for y in range(1,7) if x == 4 or y == 4]
print(res)
```

```python
# two dices combination sum convert into 1's and 0's
#even num - 0
#odd num - 1
res = [0 if(x+y)%2==0 else 1 for x in range(1,7) for y in range(1,7)]
print(res)
```

```python
#count of all combination and which is k !=(sum of three dices)
k = int(input())
res = [(x,y,z) for x in range(1,7) for y in range(1,7) for z in range(1,7) if (x+y+z) != k]
print(len(res))
```

```python
#count of all combination and which is k !=(sum of three dices)
k = int(input())
res = sum([1 for x in range(1,7) for y in range(1,7) for z in range(1,7) if (x+y+z) != k])
print(res)
```
