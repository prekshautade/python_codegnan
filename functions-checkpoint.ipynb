```python
# 1) pass by value : in pass by value the function call takes a copy of variable after executing of function defination original value will not effect it creates a new local object in function defination
# all immutable data types act as pass by value (example : string , tuple , int , floatdef fun(x):
def fun(x):    
    x = x + 1
    return x
val = 10
print(fun(val))
print(val)
    
```

```python
# 2) pass by reference : in pass by reference function call takes reference of an object (variable) after executing of function defination original value will effect
# all mutable data types (list , set , dictionary) act as pass by reference 
def fun(l,val):
    l.append(val)
    return l
lst = [1,2,3]
val = 10
print("Function returns result;",fun(lst,val))
print("Original value:",lst)
```

```python
# Recursion : a function call itself during its execution is called recursion
# the recursion allows a problem to be solved by breaking it down into smaller , similar sub problems
# def fun_name(parameter):
   # base condition 
   # recursive call
# base case or base condition : it is used to terminates from the recursive call if it skips the function execution goes to infinite loop exeution 
# recursive call or recursive case : in recursive call the function executes again and again until unless it reaches to base condition
# factorial of a number in recursion
def fact(n):
    if n == 0 :
        return 1
    else:
        return n * fact(n-1) # recursive call
print(fact(5))    

# n = 5
# 5 * fact(4)=24 # recursive case
# 4 * fact(3)=6 # recursive case
# 3 * fact(2)=2 # recursive case
# 2 * fact(1)=1 # recursive case
# 1 * fact(0)=1 # base case
# return 1
```

```python
# sorted() - it sorts the sequence
lst = ['Apple', 'Ball' , 'Cat', 'Dog']
print(sorted(lst))
print(sorted(lst, reverse = True))
print(sorted(lst , key = len))
print(sorted(lst , key = len , reverse = True))
```

```python
# enumerate() - It returns a sequence as list of tuple of index and value pairs
lst =  ['Apple', 'Ball' , 'Cat', 'Dog']
print(enumerate(lst))
print(list(enumerate(lst)))
```

```python
# any() - This function returns true if any one element in sequence is non-zero or non-empty sequence
lst = [1,2,'Hi',1,0,0,0]
print(any(lst))
print(any([0,0,0,0,0,0,0]))
print(any([1,2,0,0,0,0,0]))
print(any(['Hi',0,0,0,0,0]))
print(any(['Hi','Hi','Hi','Hi']))
```

```python
# All() - This function returns True if all elements in sequence is non-zero or non-empty sequence
lst = [1,2,'Hi',1,0,0,0]
print(all(lst))
print(all([0,0,0,0,0,0,0]))
print(all([1,2,0,0,0,0,0]))
print(all([1,2,3,4]))
print(all([1,2,3,""]))
```

```python
# map(fun,seq) - It applies the function on each element in seq
def add_one(x):
    return x + 1
lst = [10,20,30,40]
print(list(map(add_one,lst)))
```

```python
# filter(fun,seq) - It filters the sequence of element based on function
def even_numbers(x):
    if x % 2 == 0:
       return x 
lst = [10,60,301,40]
print(list(map(even_numbers,lst)))
print(list(filter(even_numbers,lst)))
```

```python
## find the square of a numbers in a list using lambda and map
lst = [1,2,3,4,5]
# map(fun,iterable)
res_list = list(map(lambda x : x**2,lst))
print(res_list)
```

```python
## filter even numbers in a list using lambda and filter
lst = [1,2,3,4,5]
# map(fun,iterable)
res_list = list(filter(lambda x : x%2==0,lst))
print(res_list)
```

```python
## Add 1 to numbers if numbers is even, otherwise add 2 to number in a list using lambda and map
lst = [1,2,3,4,5]
# map(fun,iterable)
res_list = list(map(lambda x : x+1 if x%2==0 else x+2,lst))
print(res_list)
```

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

print(count_vowels("Programming is fun!"))
```

```python
# 4. greater commom division(GCD)
n1 = 48
n2 = 18
min_no = n1 if n1 > n2 else n2
gcd = 1
for i in range(1, min_no + 1):
    if n1 % i == 0 and n2 % i == 0:
        gcd = i
print(gcd) 
```
